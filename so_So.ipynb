## Project: Data Visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities=pd.read_csv('./Data/California_cities.csv')
cities.head()
cities["area_total_km2"].sort_values()
```

```python
# Extract latd & Longd
lat, lon = cities['latd'],cities['longd']
popu, area= cities['population_total'], cities['area_total_km2']

plt.figure(figsize=(6,6))
# Plot using Pyplot API
plt.scatter(lon,lat,
           c=np.log10(popu), cmap='viridis',
           s=area, linewidths=0, alpha=0.5)
plt.axis('equal');
plt.xlabel('longlatitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}${popu}')
plt.clim(3,7);

#Create a legend for cities'sizes
area_range=[50, 100, 300]
for area in area_range:
    plt.scatter([],[],s=area, alpha=0.5,
                label=str(area)+'km$^2$')
plt.legend(labelspacing=1, title='Cty Area')

plt.title('California Cities: Population and Area Distribution');
```
